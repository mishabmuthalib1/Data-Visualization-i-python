```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

#Data Collection

```python
# Load Dataset from csv to pandas
breast_cancer_data = pd.read_csv('/content/data.csv')
```

```python
# Encode Diagnosis Column
label_encode = LabelEncoder()
labels = label_encode.fit_transform(breast_cancer_data['diagnosis'])
breast_cancer_data['target'] = labels
breast_cancer_data.drop(columns='diagnosis', axis=1, inplace=True)
```

```python
breast_cancer_data.drop(columns='id', axis=1, inplace=True)
```

**Outliers Detection**


Box plot for visualizing the outliers in the dataset

```python
for column in breast_cancer_data:
  plt.figure()
  breast_cancer_data.boxplot([column])
```

as we can see here that ost of the skewed features have Outliers

Correlation Matrix

```python
correlation_matrix=breast_cancer_data.corr()
```

```python
#constructing a heat map to visuaize the correlation matrix
plt.figure(figsize=(10,10))
sns.heatmap(correlation_matrix,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
plt.savefig('Correlation Heat Map')
```

**Multicollinearity Problem:**

Multicollinearity exists when an independent variables is highly correlated with one or more independent variables

we can remove the features if they have +ve or -ve correlation between them
